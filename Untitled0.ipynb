```
def linearsearch(a,tarItem):
  flag = 0;
  for i in range(len(a)):
    if a[i]==tarItem:
      flag=1;
    break
    if(flag!=0):
      print("target item is found")
    else:
      print("target item is not found")
      
a=[16,2,12,6,9,7,1]
linearsearch(a,61)
    
                
```

###Binary search Algorithm 
  can be applied only on sorted list
  can be applied on unique list
  

```
def binarysearch(a,leftindex,rightindex,taritem):
  while leftindex<=rightindex:
      midindex=leftindex+(rightindex-leftindex)//2;
  if a[midindex]==taritem:
    return midindex
  if a[midindex]>taritem:
      return rightindex==midindex-1
  else:
    leftindex=midindex+1
return -1
list1=[1,4,9,15,25,45,57,88,98]
res=binarysearch(list1,0,8,818)
if res!=-1:
  print("item found")
  else:
    print("item not found")
 
```

### Bubble sort 


```
def bubblesort(a):
  for i in range(len(a)-1):
    for j in range(len(a)-1):
      if a[j]>a[j+1]:
        a[j],a[j+1]=a[j+1],a[j]
     
    for i in range(len(a)):
      print(a[i],end=" ")
      
list1=[19,1,25,6,18,3]
bubblesort(list1)
```
